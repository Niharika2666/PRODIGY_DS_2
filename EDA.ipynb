```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('train.csv')
data=data.dropna()
data.head()
```

```python
data.count()
```

```python
count=0
count1=0
for i in np.array(data['Survived']):
    if i==1:
        count+=1
    else:
        count1+=1
print(count,count1)
l=[count,count1]
l1=['Survived','notsurvived']
plt.pie(l,labels=l1)
plt.savefig('piechart.png', dpi=300, bbox_inches='tight')
```

```python
plt.scatter(np.array(data['Age']),np.array(data['Survived']))
plt.title('Age vs survived')
plt.xlabel('Age')
plt.ylabel('survived')
plt.savefig('scatterplot.png', dpi=300, bbox_inches='tight')
```

```python
l=[count,count1]
l1=['Survived','notsurvived']
plt.bar(l1,l)
plt.title('survived and not survived')
plt.ylabel('count')
plt.savefig('bar.png', dpi=300, bbox_inches='tight')
```

```python
malesurvived=data[(data['Sex']=='male') & (data['Survived']==1)].shape[0]
femalesurvived=data[(data['Sex']=='female') & (data['Survived']==1)].shape[0]
m=[malesurvived,femalesurvived]
print(m)
```

```python
malenotsurvived=data[(data['Sex']=='male') & (data['Survived']==0)].shape[0]
femalenotsurvived=data[(data['Sex']=='female') & (data['Survived']==0)].shape[0]
x=['male','female']
m1=[malenotsurvived,femalenotsurvived]
width=0.25
xindex=np.arange(len(x))
print(m1)
```

```python
plt.bar(xindex-width,m,width=width,label='survived')
plt.bar(xindex,m1,width=width,label='not survived')
plt.xticks(ticks=xindex,labels=x)
plt.legend()
plt.savefig('comparision.png', dpi=300, bbox_inches='tight')
```

```python
plt.bar(xindex,m,width=width,label='survived')
plt.xticks(ticks=xindex,labels=x)
plt.title('survived vs gender')
plt.savefig('compare.png', dpi=300, bbox_inches='tight')
```
